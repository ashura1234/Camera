```python
import numpy as np
import matplotlib.pyplot as plt
import json
import scipy 
import cv2
from scipy import ndimage
import open3d as o3d
import os, os.path
```

```python
dataPath = "./data/"
rawImgPath = "./rawImage/"
depthPath = "./depth/"
imgpath = "./image/"
```

```python
dataCount = len([name for name in os.listdir('./data/')])
```

```python
for idx in range(1, dataCount + 1):
    img = cv2.imread(rawImgPath + "image" + str(idx) + ".png", cv2.IMREAD_COLOR)
    #res = cv2.resize(img, dsize=(768, 576), interpolation=cv2.INTER_CUBIC)
    pixels = np.array(img)
    res = ndimage.rotate(pixels, 90)
    cv2.imwrite(imgPath + "image" + str(idx) + ".png", res)
```

```python
for idx in range(1, dataCount + 1):
    with open(dataPath + "data" + str(idx) + ".json") as f:
        data = json.load(f)
    array = np.array(data['data'])
    mat = np.reshape(array, (-1, 768))
    print(type(mat[0][0]))
    cv2.imwrite(depthPath + "depth" + str(idx) + ".png", mat)
```

```python
with open(dataPath + "data" + str(19) + ".json") as f:
    data = json.load(f)
array = np.array(data['data'])
print(array.min())
print(array.max())
img = cv2.imread(depthPath + "depth" + str(19) + ".png", cv2.IMREAD_UNCHANGED)
print(img.min())
print(img.max())
```
