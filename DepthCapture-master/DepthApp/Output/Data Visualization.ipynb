```python
import numpy as np
import matplotlib.pyplot as plt
import json
import scipy 
import cv2
from scipy import ndimage
import open3d as o3d
```

```python
with open('Data.json') as f:
data = json.load(f)
```

```python
array = np.array(data['data'])
```

```python
mat = np.reshape(array, (-1, 768))
rotatedMat = ndimage.rotate(mat, -90)
img = cv2.imread("Photo.png", cv2.IMREAD_COLOR)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
res = cv2.resize(img, dsize=(768, 576), interpolation=cv2.INTER_CUBIC)
pixels = np.array(res)
print(np.shape(pixels))
rotatedImg = ndimage.rotate(pixels, -90)
```

```python
fig = plt.figure(figsize = (40, 60))


fig.add_subplot(1, 2, 1)
cmap_reversed = plt.cm.get_cmap('hot_r')
plt.imshow(rotatedMat, cmap_reversed)

fig.add_subplot(1, 2, 2)
plt.imshow(rotatedImg)
```
